```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Exploratory Data Analysis

```python
df=pd.read_csv('../data/HR_EMPLOYEE_CHURN_DATA.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
print(f'dataset has {df.shape[0]} number of rows \nand {df.shape[1]} number of columns ')
```

```python
df.describe()
```

```python
df.isna().sum()
df.dropna(axis=0,inplace=True,subset=['satisfaction_level'])
```

```python

fig,axis=plt.subplots(nrows=len(df.columns[1:-4].tolist()),ncols=2,figsize=(15,20))

for c,i in enumerate(df.columns[1:-4]):
    
    axis[0,0].set_title('Distribution plot \nFor \nNumerical data',
                        pad=30,
                        family= 'Arial',
                        size=16,
                        bbox=dict(mutation_aspect=0.01,boxstyle="square",ec="gray",pad=7,color='#98AFC7'))
    
    sns.distplot(df[i],ax=axis[c,0])
    
    axis[0,1].set_title('Boxplot of Numerical data \nAgainst \nEmployee left',
                        pad=30,
                        family= 'Arial',
                        size=16,
                       bbox=dict(mutation_aspect=0.01,boxstyle="square",ec="gray",pad=7,color='#98AFC7'))
    
    sns.boxplot(x=df['left'],y=df[i],ax=axis[c,1])
```

```python
columns=df.columns[-4:-1]
fig,axis=plt.subplots(nrows=len(columns.tolist()),ncols=2,figsize=(15,20))

for c,i in enumerate(columns):
    axis[0,0].set_title('Bar plot \nFor \nCategorical data',
                        pad=30,
                        family= 'Arial',
                        size=16,
                        bbox=dict(mutation_aspect=0.01,boxstyle="square",ec="gray",pad=7,color='#98AFC7'))
    
    sns.countplot(df[i],ax=axis[c,0])
    
    axis[0,1].set_title('Barplot of Categorical data \nAgainst \nEmployee left',
                        pad=30,
                        family= 'Arial',
                        size=16,
                       bbox=dict(mutation_aspect=0.01,boxstyle="square",ec="gray",pad=7,color='#98AFC7'))
    
#     sns.countplot(df[i],ax=axis[c,1])
    sns.barplot(x='left',y=i,data=df,ax=axis[c,1])
```
